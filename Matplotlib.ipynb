```python
from  matplotlib import pylab
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,25)
y=x*x+2
print(x)

print(y)
```

```python
pylab.plot(x,y,'r')
```

```python
pylab.subplot(2,2,1)  #roes colmn of fig
pylab.plot(x,y,'r--')
pylab.subplot(2,2,2)
pylab.plot(y,x,'g*-')
pylab.subplot(2,2,3)
pylab.plot(y,x,'y--')
pylab.subplot(2,2,4)
pylab.plot(y,x,'b--')
```

```python
fig=plt.figure()
axis=fig.add_axes([0.5,1,1.5,0.7])  ##left top/bottom  width height
axis.plot(x,y,'r')
```

```python
fig,axis=plt.subplots(1,3)
for ax in axis:
    ax.plot(x,y,'r')
```

```python
#one graph inside another
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.7,0.7])
axis2=fig.add_axes([0.2,0.3,0.2,0.3])
axis1.plot(x,y,'r')

axis1.set_title('axis1')
axis2.set_title('axis2')


axis2.plot(y,x,'g')




```

```python
fig=plt.figure(dpi=100)
axes=fig.add_axes([0.1,0.1,0.7,0.7])
axes.set_ylabel('y')
axes.set_xlabel('x')

axes.plot(x,x**3)
axes.plot(y,x**4)


axes.legend(["y=x**3","y=x**4"],loc=4)  # loc -locaton qudrant wise


```

```python
#graph features
fig=plt.figure(dpi=100)
axes=fig.add_axes([0.1,0.1,0.7,0.7])
axes.plot(x,x+3,'r',alpha=0.5)
axes.plot(x,x+6,'g',linewidth=1.5)
axes.plot(x,x+9,'b',marker='*')
axes.grid(True)
```

```python
fig,axes=plt.subplots(1,4,dpi=100)
axes[0].set_title("Scatter")
axes[0].scatter(x,x*4)

axes[1].set_title("step")
axes[1].step(x,x*4)

axes[2].set_title("bar")
axes[2].bar(x,x*2,width=0.5,align='center')
            

axes[3].set_title("Fill-betwwen")
axes[3].fill_between(x,x*2)            
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,1,1],polar=True)
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t,lw=5)
```

```python
fig,axes=plt.subplots(1,2,dpi=100)

axes[0].set_title("histogram")
axes[0].hist(np.random.randn(1000))

axes[1].set_title("cumulative")
axes[1].hist(np.random.randn(1000),cumulative=True,bins=5)
```

```python
#contour image

delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)

z1=np.exp(-X**2-Y**2)
z2=np.exp(-(X-1)**2-(Y-1)**2)
Z=(z1-z2)*2
```

```python
fig,ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS,inline=1)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig=plt.figure(figsize=(16,16))
ax=fig.add_subplot(1,2,1,projection='3d')
ax.plot_surface(X,Y,Z,rstride=4)
```

```python
#piechart
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y,labels=mylabels,explode=myexplode)
plt.legend()
plt.show()
```
