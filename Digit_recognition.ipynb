```python
from sklearn.datasets import load_digits
digit=load_digits()
```

```python
digit.data.shape  #image 8*8 pixel
```

```python
digit.target.shape    #label
```

```python
#display image of digits

import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,4))
for index, (image,label) in enumerate(zip(digit.data[0:5],digit.target[0:5])):
    plt.subplot(1,5,index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title("training %i\n" %label)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digit.data,digit.target,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train,y_train)
```

```python
pred=LR.predict(x_test)
```

```python
score=LR.score(x_test,y_test)
score
```

```python
from sklearn import metrics
import seaborn as sns
plt.figure(figsize=(16,9))
cm=metrics.confusion_matrix(y_test,pred)
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
sns.heatmap(cm,annot=True)
```
